```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("netflix_dataset_50.csv")
```

```python
df.head(2)
```

# count of each genre

```python
genre_count = df.groupby("genre").size().reset_index(name="count")
genre_count
```

# each genre avg rating

```python
genre_avg_rating = df.groupby("genre")["rating"].mean().reset_index(name="avg_rating")
genre_avg_rating
```

# year wise trand on genre

```python
genre_pr_year = df.groupby(['release_year', 'genre']).size().reset_index(name='count')
genre_pr_year
```

# Most popular and letest genre identfy

```python
genre_counts = df['genre'].value_counts()

most_popular = genre_counts.idxmax()       
most_popular_count = genre_counts.max()    

# Least popular genre
least_popular = genre_counts.idxmin()      
least_popular_count = genre_counts.min()   

print(f"Most popular genre: {most_popular} ({most_popular_count} occurrences)")
print(f"Least popular genre: {least_popular} ({least_popular_count} occurrences)")
```
